### 틱(tick) 설정
- 축에 위치를 표시하는 점을 틱(tick) 이라고 한다.

```python
x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
plt.title('x축과 y축 tick 설정')
plt.plot(x,c)
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
plt.title('x축과 y축 tick 설정')
plt.plot(x,c)
plt.xticks([-np.pi,-np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
plt.title('x축과 y축 tick 설정')
plt.plot(x,c)
plt.xticks([-np.pi,-np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$',r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,1],['low','zero','high'])
plt.show()
```

### 범례(legend)
- 하나의 차트에 여러개의 그래프를 그릴경우 각 그래프가 뜻하는 것을 표시하는 것
- legend()
- loc : best(0)
- loc : upper right(1)
- loc : upper left(2)
- loc : lower left(3)
- loc : lower right(4)
- loc : right(5)
- loc : center left(6)
- loc : center right(7)
- loc : lower center(8)
- loc : upper center(9)
- loc : center(10)

```python
x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
s = np.sin(x)
plt.title('범례를 가진 그래프')
plt.plot(x, c, ls='--', label='cosine')
plt.plot(x, s, ls=':', label='sine')
plt.legend(loc=2)
plt.show()
```

**Axes객체와 subplot()**
- http://matplotlib.org/api/axes_api.html#matplotlib.axes.Axes
- subplot(2,1,1)

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0,2.0)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)
```

```python
ax1 = plt.subplot(2,2,1)
plt.plot(x1,y1,'yo-')
plt.title('첫번째 subplot')
plt.ylabel('A')

ax2 = plt.subplot(2,2,2)
plt.plot(x2,y2,'r.-')
plt.xlabel('time')
plt.ylabel('B')

ax3 = plt.subplot(2,2,3)
plt.plot(x1,y1,'yo-')
plt.title('세번째 subplot')
plt.ylabel('C')

ax4 = plt.subplot(2,2,4)
plt.plot(x2,y2,'r.-')
plt.xlabel('etc')
plt.ylabel('D')

plt.tight_layout() # 플롯간의 간격을 자동으로 맞춰줌
plt.show()
```

**Axis 객체와 축**
- 하나의 Axes에 두개이상의 Axis를 가진다.
- https://matplotlib.org/api/axis_api.html
- twinx() 메서드를 이용하여 복수개의 축을 설정한다.

```python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()

ax0.set_title('2개의 y축을 가진 하나의 figure 객체')
ax0.plot([10,5,2,9,7], 'r-', label ='y0')
ax0.set_ylabel('y0')
ax0.grid(True)

ax1.plot([100,200,220,180,120], 'g:', label = 'y1')
ax1.set_ylabel('y1')
ax1.grid(True)

ax0.set_xlabel('공유하는 x축')
plt.show()
```

```python
fig, ax = plt.subplots(3,4, figsize=(15,10))

for row in range(3):
    for col in range(4):
        ax[row, col].plot(np.random.randn(10))
        ax[row, col].set_title(f'{row+1}행 {col+1}열')
        ax[row, col].set_xlabel(f'{row+1}행')
        ax[row, col].set_ylabel(f'{col+1}열', rotation=0)
fig.tight_layout()
plt.show()
```

```python
y = [2,3,1]
x = np.arange(len(y))
xlabel = ['가','나','다']
plt.title('bar chart')
plt.bar(x,y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel('가나다')
plt.ylabel('빈도수')
plt.show()
```

```python
y = [2,3,1]
x = np.arange(len(y))
xlabel = ['가','나','다']
plt.title('bar chart')
plt.barh(x,y, alpha =0.4)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel('가나다')
plt.ylabel('빈도수')
plt.show()
```

**stem plot**
- 바차트와 비슷하지만 폭이 없는 그래프
- 주로 이산확률함수나 자기상관관계를 나타낼때 사용
-https://matplotlib.org/2.0.2/api/pyplot_api.html?highlight=stem%20plot#matplotlib.pyplot.stem

```python
x = np.linspace(0.1, 2*np.pi, 10)
y = np.cos(x)

markerline, stemlines, baseline = plt.stem(x, np.cos(x), '-.')
plt.setp(baseline, 'color', 'r', 'linewidth', 2)

plt.show()
```

## 파이차트
- pie()
- 파이차트를 그릴때 원의 형태를 유지하도록 plt.axis('equal')명령

```python
labels=['개구리','돼지','개','통나무']
sizes = [15,30,45,10]
colors = ['yellowgreen', 'gold', 'lightskyblue','lightcoral']
explode = (0,0.1,0,0)

plt.title('pie chart')

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%')
plt.show()
```

## 스캐터 플롯(scatter plot)

```python
np.random.seed(0)
x = np.random.normal(0,1,100)
y = np.random.normal(0,1,100)

plt.title('scatter plot')
plt.scatter(x,y)

plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
N=30
np.random.seed(0)
x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi*(15*np.random.rand(N))**2
plt. title('bubble chart')
plt.scatter(x,y1, c=y2, s =y3)
```
