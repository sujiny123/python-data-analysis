- 인증된 url만 데이터에 접근할 수 있는 권한을 부여하는 방식


```python
# JSON 데이터를 다루는 방법
# 파이썬의 딕셔너리와 리스트가 중첩된 데이터 형태를 JSON 이라고 부른다.
dic = {'name':'홍길동'}
dic['name']
```

```python
import json
```

```python
# ensure_ascii=False : 한글을 포함하고 있을 때, 원래 저장된 문자를 그대로 출력하도록 지정하는 옵션
dic_str = json.dumps(dic, ensure_ascii=False)
print(dic_str)
```

# 파이썬을 통한 XML 파일 다루는 작업
Extensible Markup Language : XML
HyperText Markup Language : HTML

```python
x_str = """
<book>
    <name>코딩도장</name>
    <author>남재윤</author>
    <year>2022</year>
</book>
"""
```

```python
import xml.etree.ElementTree as et
```

```python
book = et.fromstring(x_str)
```

```python
# root tag = 부모태그
book.tag
```

```python
book_childs = list(book)
print(book_childs)
```

```python
# unpacking : 하나의 변수로 묶여있는 값을 각각의 변수로 변환하는 작업
name, author, year = book_childs
```

```python
print(name.text)
```

```python
# findtext()
name = book.findtext('name')
```

```python
name
```

```python
author = book.findtext('author')
year = book.findtext('year')
```

```python
x2_str = """
<books>
    <book>
          <name>코딩도장</name>
          <author>남재윤</author>
          <year>2022</year>
    </book>
    <book>
         <name>데이터시각화</name>
          <author>설지윤</author>
          <year>2023</year>
   </book>
</books>
"""
```

```python
books = et.fromstring(x2_str)
print(books.tag)
```

```python
# 여러개의 자식태그를 확인하여 읽어오는 작업
# 모든 자식 객체를 반환하는 메서드 findall()
for book in books.findall('book'):
  name = book.findtext('name')
  author = book.findtext('author')
  year = book.findtext('year')
  print(name, author, year, '\n')
```

```python
soup = BeautifulSoup(r.text,'html.parser')
detail = soup.find('div', attrs={'id':'infoset_specific'})
print(detail)
```

```python
tr_list = detail.find_all('tr')
```

```python
tr_list
```

```python
#태그안에 텍스트를 가져오는 작업 : get_text()
for tr in tr_list:
  if tr.find('th').get_text() == '쪽수, 무게, 크기':
    data = tr.find('td').get_text()
    break
```
